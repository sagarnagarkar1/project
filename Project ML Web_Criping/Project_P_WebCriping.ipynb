```python
#load the library
import requests
# Add Link
Wiki_link="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
link=requests.get(Wiki_link).text
```

```python
print(link)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(link,'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.find_all("a")
```

```python
all_link=soup.find_all("a")
for link in all_link:
      print(link.get("href"))
```

```python
all_tables=soup.find_all('table')
print(all_tables)
```

```python
right_table=soup.find('table',class_="wikitable sortable")
right_table
```

```python
table_links=right_table.findAll('a')
table_links
```

```python
country =[]
for links in table_links:
      country.append(links.get('title'))
print(country)
```

```python
import pandas as pd
pruthvi=pd.DataFrame()
pruthvi['Country']=country
pruthvi
```
